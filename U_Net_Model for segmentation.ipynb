Data preparation

```python
from arcgis.learn import UnetClassifier, prepare_data
```

```python
#On this step, 20% of data is for validation and 80% for training
data = prepare_data(path=r'D:\informal\Image_classification\Training_samples\Training', chip_size=128, batch_size=4)
```

```python
data.show_batch(rows=3)
```

Load model architecture

```python
unet_model = UnetClassifier(data, backbone='resnet34', ignore_classes=[0])
```

```python
unet_model.unfreeze()
```

```python
#Observe the behaviour of untrained model
unet_model.show_results(rows=5, alpha=1)
```

```python
#Finding the optimal learning rate
unet_model.lr_find()
```

Training the model

```python
unet_model.fit(100, lr=0.0001)
```

```python
#Ploting the learning curve
unet_model.learn.recorder.plot_losses()
```

```python
#Preview the model results on validation set
unet_model.show_results()
```

Saving the model for inferencing

```python
unet_model.save('Unet_model_trained_200')
```
